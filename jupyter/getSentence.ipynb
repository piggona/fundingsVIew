## 读XML文件，提取基金描述

```python
import xml.dom.minidom
from xml.etree.ElementTree import parse
import os

filePath = '/Users/haohao/Documents/fundingsView/data/raw/'
```

### 获得文件夹下的所有文件名称，并将文件处理

```python
os.listdir(filePath)
```

```python
def getAbstract(fileName):
    doc = parse(filePath+file)
    root = doc.getroot()
    childs = root.getchildren()

    abstract = ""
    for child0 in childs:
        for child00 in child0.getchildren():
            # print child00.tag #标签名，即name、date、price、description
            # print child00.text
            if child00.tag == "AbstractNarration":
                abstract = child00.text

    return abstract
```

```python
xmllist = os.listdir(filePath)
file = xmllist[0]
xml = file + filePath
abstract = getAbstract(xml)
print(abstract)
```
