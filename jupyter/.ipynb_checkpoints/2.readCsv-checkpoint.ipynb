# 导入打标数据

## 读csv

```python
import pandas as pd

tag_data = pd.read_csv('/Users/haohao/Documents/fundingsView/jupyter/file2.csv')
print(tag_data)
```

## 统计数据

```python
tag_data = tag_data[['label','text']]
print(tag_data.groupby(['label']).count())
```

## 替换标记

```python
tag_data = tag_data.query('label==3 | label ==4')
print(tag_data)
```

```python
def standard_label(x):
    if x == 3:
        return 1
    elif x == 4:
        return 0
```

```python
tag_data["label"] = tag_data["label"].apply(standard_label)
print(tag_data)
```

### 去掉标点符号及html符

```python
import re
def removeDots(x):
    string = re.sub("[\s+\.\!\/_,$%^*(+\"\']+|[+——！，。？、~@#￥%……&*（）]+", " ",x)
    p = re.compile('<.*?>')
    return p.sub('',string)
```

```python
tag_data["text"] = tag_data["text"].apply(removeDots)
print(tag_data)
```

```python
print(type(tag_data))
```

```python
numpy_data = tag_data.values
print(numpy_data)
```

## 存为pickle

```python
import pickle

output = open('labeled_data.pkl','wb')

pickle.dump(numpy_data, output)

output.close()
```
